```python
def wordopt(text):
    text= text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub("\\W"," ",text)
    text=re.sub('https?://\S+|www\.\S+','',text)
    text=re.sub('<.*?>+','',text)
    text=re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text=re.sub('\n','', text)
    text=re.sub('\W*\d\w*','',text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['label']
```

```python
x_train, x_test ,y_train, y_test =train_test_split(x,y, test_size=0.25)
```

```python
vbectorization=TfidfVectorizer()
xv_train=vbectorization.fit_transform(x_train)
xv_test= vbectorization.transform(x_test)
```

```python
lr=LogisticRegression()
lr.fit(xv_train,y_train)
```

```python
pred_lr=lr.predict(xv_test)
lr.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb= GradientBoostingClassifier(random_state=0)
gb.fit(xv_train,y_train)
```

```python
def output_lable(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "Real News"
```

```python
import pyttsx3
import tempfile
import os
```

```python


# Assuming 'wordopt', 'vbectorization', 'lr', and 'output_label' are already defined

def manual_testing(news):
    testing_news = {"text": [news]}
    new = pd.DataFrame(testing_news)
    new["text"] = new["text"].apply(wordopt)
    new_x_test = new["text"]
    new_xv_test = vbectorization.transform(new_x_test)
    pred_lr = lr.predict(new_xv_test)
    result = output_lable(pred_lr[0])

    # Initialize TTS engine
    engine = pyttsx3.init()
    engine.setProperty('rate', 150)  # Adjust speech rate if desired

    # Create a temporary file to save the audio
    with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
        audio_path = fp.name
    engine.save_to_file(f"The news is {result}", audio_path)
    engine.runAndWait()

    return result, audio_path

# Define Gradio interface
iface = gr.Interface(
    fn=manual_testing,
    inputs=gr.Textbox(lines=2, placeholder="Enter news text here..."),
    outputs=[
        gr.Textbox(label="Prediction"),
        gr.Audio(label="Spoken Output", autoplay=True)
    ],
    title="Fake News Detection",
    description="Enter news text to check if it's fake or real. The result will also be spoken aloud."
)

iface.launch(debug=True)
```
